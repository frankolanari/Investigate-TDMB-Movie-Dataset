> #### Function 3 : Seperate and fetch the value count of  individual values in a colum
In order to easily answer the subsequent question efficiently with less ambiguity,  we'll define a function that seperate and  fetch out strings in a column seperated by vertical line ( | ). The function would allow me choose a ***column*** of interest from a selected ***dataframe*** to work on and also allow me choose the ***number*** of result I want. I made it flexible, cause I might need to use this function later on.

```python
#Defining the function to seperate and fetch the  top value counts... 
def fetch_vcount(dataframe,column, number):
    #This seperates the values by the '|' seprator
    col_values = dataframe[column].str.cat(sep = '|')
    #storing them in a pandas series
    col_values = pd.Series(col_values.split('|'))
    #getting the value counts 
    v_counter = col_values.value_counts(ascending = False)
    #determing the how many of the highest value counts to return
    v_count = v_counter.head(number)
    return v_count
```

>##### Function 4 : Return Highest and Lowest Values in a Column
Inorder to make our code less ambigous, Let's define a function that returns and concatenates the highest and lowest values in a column. This would make answering the first four basic questions quite easy and done with simpler code.

```python
##Defining a function that calculates the highest and lowest value in a column
def Cal_MinMax(dataframe,column):
    #calculates  the highest value in the column of interest
    maxv = dataframe[column].idxmax()
    max_value = pd.DataFrame(topmovies.loc[maxv])
    
    #Calculates the lowest value in the column of interest
    minv = dataframe[column].idxmin()
    min_value = pd.DataFrame(topmovies.loc[minv])
    
    #Concatinating our max and min values
    concat_MinMax = pd.concat([max_value, min_value], axis = 1)
    
    return concat_MinMax
```
